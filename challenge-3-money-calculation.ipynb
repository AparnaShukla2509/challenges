# Money calculation

**Difficulty level**: 1 - beginner

Imagine you need to create a calculation that, for a given amount of money, it will return the number of each 
bill and coin you need to reach that amount. The priority order for the bills is from the highest bill to the smallest coin.

Assumptions:

- Available bills: `100$`, `50$`, `20$`, `10$` and `5$`
- Available coins: `1$`, `0.5$`, `0.1$`, `0.05$` and `0.01$`
- The largest bill/coin has priority compared to the smallest one.
- The return of the function should be a dictionary, where the keys are represented by the nominal value of the 
  bill/coin and the values are represented bty the number of bills/coins needed.
- Float is not a precise data format, use `decimal.Decimal` instead.
  
Tips:

- Some functions that you will need for this challenge: 
  - `for` loop.
  - the dictionary `keys` method.
  - string slicing.
  - casting from string to decimal using `decimal.Decimal` method.
  - for a division with no rest use `//`.
  - cast a decimal or float number to integer using `int` method.
